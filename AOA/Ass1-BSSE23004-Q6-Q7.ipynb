```python
# =======================
#   Q6
# =======================

class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = []          # list of tuples: (weight, u, v)
        self.mst = []            # list of MST edges
        self.mst_cost = 0

    def add_node(self, node):
        if node in self.nodes:
            print(f"Node {node} already exists.")
            return
        self.nodes.add(node)
        self.rebuild_mst()
        print(f"Node {node} added.")

    def add_edge(self, u, v, weight):
        if u == v:
            print("Self-loops are not allowed.")
            return
        if u not in self.nodes or v not in self.nodes:
            print("Both nodes must exist before adding an edge.")
            return

        # Remove old edge if it already exists (undirected)
        key = tuple(sorted((u, v)))
        self.edges = [
            (w, a, b) for (w, a, b) in self.edges
            if tuple(sorted((a, b))) != key
        ]

        self.edges.append((weight, u, v))
        self.rebuild_mst()
        print(f"Edge ({u}, {v}, {weight}) added.")

    def delete_node(self, node):
        if node not in self.nodes:
            print(f"Node {node} does not exist.")
            return

        self.nodes.remove(node)
        self.edges = [
            (w, u, v) for (w, u, v) in self.edges
            if u != node and v != node
        ]
        self.rebuild_mst()
        print(f"Node {node} deleted.")

    def rebuild_mst(self):
        self.mst = []
        self.mst_cost = 0

        parent = {node: node for node in self.nodes}
        rank = {node: 0 for node in self.nodes}

        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])  # Path compression
            return parent[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx == ry:
                return False

            # Union by rank
            if rank[rx] < rank[ry]:
                parent[rx] = ry
            elif rank[rx] > rank[ry]:
                parent[ry] = rx
            else:
                parent[ry] = rx
                rank[rx] += 1
            return True

        for weight, u, v in sorted(self.edges, key=lambda x: x[0]):
            if u in self.nodes and v in self.nodes and union(u, v):
                self.mst.append((weight, u, v))
                self.mst_cost += weight

    def mst_nodes(self):
        nodes_in_mst = set()
        for _, u, v in self.mst:
            nodes_in_mst.add(u)
            nodes_in_mst.add(v)
        return nodes_in_mst

    def search_node_in_mst(self, node):
        if node in self.mst_nodes():
            print(f"Node {node} exists in MST.")
        else:
            print(f"Node {node} does not exist in MST.")

    def search_edge_in_mst(self, u, v):
        key = tuple(sorted((u, v)))
        for w, a, b in self.mst:
            if tuple(sorted((a, b))) == key:
                print(f"Edge ({u}, {v}) exists in MST with weight {w}.")
                return
        print(f"Edge ({u}, {v}) does not exist in MST.")

    def display_mst(self):
        if not self.mst:
            print("MST is empty or graph is disconnected.")
            print("Current MST cost:", self.mst_cost)
            return

        print("\nCurrent MST edges:")
        for weight, u, v in self.mst:
            print(f"({u}, {v}) -> {weight}")
        print("Total MST cost:", self.mst_cost)


def main():
    graph = Graph()

    while True:
        print("\n--- MENU ---")
        print("1. Add Node")
        print("2. Add Edge")
        print("3. Delete Node")
        print("4. Search Node in MST")
        print("5. Search Edge in MST")
        print("6. Display MST")
        print("7. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            node = input("Enter node label: ").strip()
            graph.add_node(node)

        elif choice == "2":
            u = input("Enter first node: ").strip()
            v = input("Enter second node: ").strip()
            try:
                weight = float(input("Enter edge weight: ").strip())
                if weight.is_integer():
                    weight = int(weight)
            except ValueError:
                print("Invalid weight.")
                continue
            graph.add_edge(u, v, weight)

        elif choice == "3":
            node = input("Enter node label to delete: ").strip()
            graph.delete_node(node)

        elif choice == "4":
            node = input("Enter node label to search in MST: ").strip()
            graph.search_node_in_mst(node)

        elif choice == "5":
            u = input("Enter first node: ").strip()
            v = input("Enter second node: ").strip()
            graph.search_edge_in_mst(u, v)

        elif choice == "6":
            graph.display_mst()

        elif choice == "7":
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Try again.")


if __name__ == "__main__":
    main()
```

```python
# =======================
#   Q7
# =======================



def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def build_max_heap(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)


def heap_sort(arr):
    n = len(arr)
    build_max_heap(arr)

    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)


# Given array
A = [5, 18, 10, 25, 30, 12, 40, 6, 8, 35, 20, 15]

print("Original Array:", A)

build_max_heap(A)
print("Max Heap:", A)

heap_sort(A)
print("Sorted Array:", A)
```
